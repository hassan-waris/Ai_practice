```python
import re

text = input()

floor = re.search(r'floor (\d+)', text).group(1)
aisle = re.search(r'aisle ([A-Z]+)', text).group(1)
box = re.search(r'box (\w+)', text).group(1)

print('Floor:', floor)
print('Aisle:', aisle)
print('Box:', box)
```
